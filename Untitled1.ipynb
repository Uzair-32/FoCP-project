```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n-cls.pt")

# Train the model
results = model.train(data="/content/Car-Model-1", epochs=5, imgsz=64)
```

```python
model = YOLO("/content/runs/classify/train2/weights/best.pt")

image_path = "/content/Test.jpg"

results = model(image_path)

# Print predictions
for result in results:
    print(f"Predicted Classess: {result.probs.top5}")
    print(f"Predicted Class: {result.names[result.probs.top1]}")  # Top-1 class
    print(f"Confidence: {result.probs.top1conf:.2f}")  # Confidence score
```
